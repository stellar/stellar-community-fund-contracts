```python
import json
from pathlib import Path

trust_file = Path("../neurons/result/users_trust_history.json")
trust = trust_file.read_text("utf-8")
trust = json.loads(trust)
```

```python
trust_collected = []

for (key, value) in trust.items():
    entry = {}
    entry["public_key"] = key
    for count, t in enumerate(value):
        entry[str(count+27)] =int(t) / 10**18 
    trust_collected.append(entry)
```

```python
import csv

with open("history_trust.csv", "w") as f:
    csvwriter = csv.DictWriter(f, fieldnames=["public_key", "27","28","29","30","31","32"])
    csvwriter.writeheader()
    csvwriter.writerows(trust_collected)
```
