```python
import json
from pathlib import Path

old_trust_file = Path("../neurons/result/trust_graph_neuron_0.json")
old_trust = old_trust_file.read_text("utf-8")
old_trust = json.loads(old_trust)
```

```python
old_trust_collected = []

for (key, value) in old_trust.items():
    entry = {}
    entry["public_key"] = key
    entry["trust"] = int(value) / 10**18 
    old_trust_collected.append(entry)
```

```python
import csv

with open("old_trust.csv", "w") as f:
    csvwriter = csv.DictWriter(f, fieldnames=["public_key", "trust"])
    csvwriter.writeheader()
    csvwriter.writerows(old_trust_collected)
```
